```python
# pylint: disable-all

import pandas as pd
import json

from time import sleep
from kafka import KafkaProducer
```

```python
producer = KafkaProducer(
    bootstrap_servers=['54.167.119.183:9092'],
    value_serializer=lambda x: json.dumps(x).encode('utf-8')
)
```

```python
data = pd.read_csv("./stock_market_data/indexProcessed.csv")
data.head()
```

```python
for i in range(50):
    dt_value = data.sample(1).to_dict(orient='records')[0]
    producer.send('kafka_test_action', value=dt_value)
    print(dt_value)
    sleep(2)
```
